```python
# importing necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.DataFrame(columns=['title','place','price','area_sqft','status'])
```

```python

for i in range(1,1000):
  url = 'https://www.makaan.com/mumbai-residential-property/rent-property-in-mumbai-city?page={}'.format(i)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get(url,headers=headers).text
  soup = BeautifulSoup(webpage,'lxml')
  container = soup.find_all('div',class_='cardLayout clearfix')
  title = []
  place = []
  price = []
  area_sqft = []
  status = []

  for i in container:
    title.append(i.find('div',class_='title-line mar-right').text)
    place.append(i.find('a',class_='loclink').text)
    price.append(i.find('td',class_='price').text)
    area_sqft.append(i.find('td',class_='size').text)
    status.append(i.find('td',class_='val').text)

    temp_df = pd.DataFrame({'title':title,'place':place,'price':price,'area_sqft':area_sqft,'status':status})
    print(temp_df)
    df = pd.concat([df,temp_df],ignore_index=True)
```
