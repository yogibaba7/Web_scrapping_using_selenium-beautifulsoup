```python
with open('smartprix.html','r',encoding='utf-8') as f:
  html = f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
soup
```

```python
container = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
container[1]
```

```python
# getting image link
img = []
for i in container:
  img.append(i.find('img',{'class':"sm-img"}))
print(len(img))
```

```python
# getting phone link
phone_name = []
for i in container:
  phone_name.append(i.find('a',{'class':'name clamp-2'}).text)

print(len(phone_name))
```

```python
# price
price = []
for i in container:
  price.append(i.find('span',{'class':'price'}).text)

print(len(price))
```

```python
# specs
rating = []
for i in container:
  j = i.find('div',{'class':'score rank-1-bg'})
  try:
    rating.append(j.find('b').text)
  except:
    rating.append(np.nan)

print(len(rating))
```

```python
yogi = soup.find_all('ul',{"class":'sm-feat specs'})
print(yogi)
```

```python
# connectibity,proccessor
conn = []
processor = []
memory = []
bettry = []
display = []
camera = []
memory_card = []
os = []
for i in yogi:
  cnt = i.find_all('li',{'class':""})
  conn.append(cnt[0])
  processor.append(cnt[1])
  memory.append(cnt[2])
  bettry.append(cnt[3])
  display.append(cnt[4])
  camera.append(cnt[5])
  memory_card.append(cnt[6])
  try:
    os.append(cnt[7])
  except:
    os.append(np.nan)
print(os)
```

```python
conn = []
processor = []
memory = []
bettry = []
display = []
camera = []
memory_card = []
os = []
```

```python
df = {
    'Phone':phone_name,
    'Img_url':img,
    'Price':price,
    'Rating':rating,
    'connectivity':conn,
    'processor':processor,
    'Memory':memory,
    'Bettry':bettry,
    'Display':display,
    'Camera':camera,
    'Memory_card':memory_card,
    'Os':os

}

webscrapping_df  = pd.DataFrame(df)
```

```python
webscrapping_df.to_csv('advance_webscarping_df.csv')
```
