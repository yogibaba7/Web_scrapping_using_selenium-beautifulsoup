```python
with open('leptop.html','r',encoding='utf-8') as f:
  html = f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
container = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
name = []
for i in container:
  name.append(i.find('h2').text)

len(name)
```

```python
price = []
for i in container:
  price.append(i.find('span',{'class':'price'}).text)

price
```

```python
rating = []
for i in container:
  try:
    rating.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
  except:
    rating.append(np.nan)

rating
```

```python
processor = []
core = []
ram = []
memory = []
graphic = []
display = []
window = []
warrenty = []
for i in container:
  try:
    processor.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[0].text)

  except:
    processor.append(np.nan)


  try:
    core.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[1].text)
  except:
    core.append(np.nan)

  try:
    ram.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[2].text)
  except:
    ram.append(np.nan)


  try:
    memory.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[3].text)
  except:
    memory.append(np.nan)


  try:
    graphic.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[4].text)
  except:
    graphic.append(np.nan)

  try:
    display.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[5].text)
  except:
    display.append(np.nan)

  try:
    window.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[6].text)
  except:
    window.append(np.nan)

  try:
    warrenty.append(i.find('ul',{'class':'sm-feat specs'}).find_all('li')[7].text)
  except:
    warrenty.append(np.nan)




len(processor)
len(core)
ram
memory
graphic
display
window
warrenty
price
```

```python
df = {'name':name,
      'price':price,
      'rating':rating,
      'processor':processor,
      'core':core,
      'ram':ram,
      'memory':memory,
      'graphic':graphic,
      'display':display,
      'window':window,
      'warrenty':warrenty

      }
```

```python
final_df = pd.DataFrame(df)
```

```python
final_df.to_excel('leptop.xlsx',index=False)
```
