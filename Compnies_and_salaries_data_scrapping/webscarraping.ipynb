```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}



df = pd.DataFrame()

for j in range(1,10):
  url = "https://www.ambitionbox.com/list-of-companies?page={}".format(j)
  web = requests.get(url,headers=headers).text

  soup = BeautifulSoup(web,"lxml")

  company = soup.find_all("div",class_="companyCardWrapper")

  name = []
  rating = []
  tags = []
  employees = []
  ctype = []
  age = []

  for i in company:
    name.append(i.find("h2",class_="companyCardWrapper__companyName").text.strip())
    rating.append(i.find("div",class_="companyCardWrapper__companyRatingWrapper").text)
    tags.append(i.find("span",class_="companyCardWrapper__interLinking").text.strip().split("|")[0])
    employees.append(i.find("span",class_="companyCardWrapper__interLinking").text.strip().split("|")[1])
    try:
      ctype.append(i.find("span",class_="companyCardWrapper__interLinking").text.strip().split("|")[2])
    except:
      ctype.append(np.nan)
    try:
      age.append(i.find("span",class_="companyCardWrapper__interLinking").text.strip().split("|")[3])
    except:
      age.append(np.nan)

  temp_df = {"Company_name":name,"Rating":rating,"Tags":tags,"Employees":employees,"Company_types":ctype,"Age":age}
  df1 = pd.DataFrame(temp_df)

  df = df.append(df1,ignore_index=True)

df
```
